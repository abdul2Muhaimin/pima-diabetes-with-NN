## Loading Libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from keras.utils import to_categorical
import matplotlib.pyplot as plt
from keras.utils import plot_model


from keras.callbacks import Callback
import keras.backend as K

import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.metrics import confusion_matrix,classification_report
```

## Loading Dataset

```python

# --- reading the dataset
df= pd.read_csv('pima-indians-diabetes.csv')
data=df.values
X = data[:,:8]
Y = data[:,8]

batch_size = 16
epochs=200
rate=0.5

```

## Train and Test Split

```python
encoded_Y = to_categorical(Y)

X_train, X_test, y_train, y_test = train_test_split(X,encoded_Y,test_size=0.3,random_state=7,shuffle=False)
```

## Normalization of Dataset

```python
scaler = MinMaxScaler()
scaled_x_train = scaler.fit_transform(X_train)
scaled_x_test = scaler.transform(X_test)

```

## Neural Network Modeling

```python
adam = keras.optimizers.Adam(lr=0.001)

model = keras.models.Sequential()
model.add(keras.layers.Dense(1, input_dim=8, kernel_initializer='uniform'))
model.add(keras.layers.LeakyReLU(alpha=0.2))
model.add(keras.layers.Dense(2,  kernel_initializer='uniform', activation='softmax'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])


# Fit the model
history=model.fit(scaled_x_train, y_train,
                  validation_data=(scaled_x_test, y_test),
                  epochs=epochs,
                  batch_size=batch_size,
                  shuffle=False)
```

## Model Evaluation

```python
scores = model.evaluate(scaled_x_test, y_test)
print()
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# summarize history for accuracy
plt.figure(1)
plt.title('accuracy')
plt.plot(history.history['accuracy'], label='training_accuracy', color = 'blue')
plt.plot(history.history['val_accuracy'], label = 'testing_accuracy', color = 'brown')
plt.legend()
plt.show()
```

```python
# summarize history for loss
plt.figure(2)
plt.title('loss')
plt.plot(history.history['loss'], label='testing_loss', color = 'blue')
plt.plot(history.history['val_loss'],label='training_loss', color = 'brown')
plt.legend()
plt.show()
```

## Model Prediction

```python
y_pred = model.predict(scaled_x_test)
print(classification_report(np.argmax(y_pred,axis=1),np.argmax(y_test,axis=1)))
print(confusion_matrix(np.argmax(y_pred,axis=1),np.argmax(y_test,axis=1)))
```
